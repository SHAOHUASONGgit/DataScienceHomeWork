## 现值计算

```python
def calculateValue(rate, cashFlow, presentYear):
    totalYear = len(cashFlow)
    output = [0] * totalYear
    for year in range(totalYear):
        output[year] = cashFlow[year] / ((1 + rate) ** (year + 1 - presentYear))
    return output
```

```python
C1 = [1] * 100
C2 = list(range(100))
C3 = [year * (1.05 ** year) for year in range(100)]
C4 = [year * (1.06 ** year) for year in range(100)]
C1 = calculateValue(0.05, C1, 0)  # k=0 时为现在年
C2 = calculateValue(0.05, C2, 1)  # k=1 时为现在年
C3 = calculateValue(0.05, C3, 1)
C4 = calculateValue(0.05, C4, 0)
print("C1:" + str(C1) + "\n")
print("C2:" + str(C2) + "\n")
print("C3:" + str(C3) + "\n")
print("C4:" + str(C4) + "\n")
```

## 有向无环图个数计算

```python
def calculateCombo(total, part):  # 组合数计算函数
    up = 1
    down = 1
    for i in range(part):
        up *= total - i
        down *= part - i
    output = up / down
    return int(output)

def calculateGraphNums(maxN, n, numList):  # maxN:顶点个数
    if maxN == 0:
        return numList
    elif n == maxN:
        sum = 0
        for k in range(1, n + 1):
            part1 = (-1) ** (k - 1)
            part2 = 2 ** (k * (n - k))
            part3 = calculateCombo(n, k)
            part4 = numList[n-k]
            sum += part1 * part2 * part3 * part4
        numList.append(sum)
        return numList
    else:
        sum = 0
        for k in range(1, n + 1):
            part1 = (-1) ** (k - 1)
            part2 = 2 ** (k * (n - k))
            part3 = calculateCombo(n, k)
            part4 = numList[n-k]
            sum += part1 * part2 * part3 * part4
        numList.append(sum)
        return calculateGraphNums(maxN, n + 1, numList)
        
```

```python
for tens in range(1, 11):
    print(str(tens * 10) + "顶点有向无环图枚举个数:" + str(calculateGraphNums(tens * 10, 1, [1])[-1]) + "\n")
```
